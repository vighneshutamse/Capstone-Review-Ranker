***In this section we will fit the Random Forest and XG Boost models only on the original features i.e. the features that we created like Length of the reviews, percentage of nouns etc which is present in the excel sheet Corrected_New_Features.csv.***

```python
# Importing Necessary Modules
import pandas as pd
import numpy as np
import spacy
nlp=spacy.load('en_core_web_sm')
from sklearn.model_selection import train_test_split
from spacy.pipeline import SentenceSegmenter
import re
import warnings
warnings.filterwarnings('ignore')
import textblob
from textblob import TextBlob
from spacy.lang.en import English

# Customization of View
pd.options.display.max_columns = 999
```

```python
# Importing the dataset containing unigram feature set
df = pd.read_csv('Unigram_Feature.csv',index_col='Sr_No')
```

```python
df.head()
```

```python
# Ignoring all the unigrams and taking only the original features
df_ = df.iloc[:,-17:]
```

```python
df_.head()
```

### Splitting into Train and Test data

```python
# All predictors in X
X=df_.iloc[:,:-1]
# The target variables 'h' in y
y=df_.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
# Splitting into train and test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=50,random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

### Random Forest Regressor: Modelling the train data

```python
# Import the model
from sklearn.ensemble import RandomForestRegressor
```

```python
# Training the Random Forest Regressor
rf = RandomForestRegressor(n_estimators = 100, random_state = 0)
rf.fit(X_train, y_train)

# Getting the feature importance
feature_importances_df = pd.DataFrame({"feature": X_train.columns, "importance": rf.feature_importances_}).sort_values("importance", ascending = False)
feature_importances_df.head(20)
```

### Predicting on Test Data

```python
# Predicting on test data
y_pred = rf.predict(X_test)
```

```python
y_pred
```

```python
# Creating a dataframe to compare predicted and test values
pd.DataFrame({"y_pred": y_pred, "y_test": y_test})
```

### Calculating Mean Absolute Percentage Error (MAPE)

```python
# Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
```

```python
mape
```

### XGBoost Regressor: Modelling the train data

```python
# Training the XG Boost Regressor
from xgboost import XGBRegressor
xgb = XGBRegressor(n_estimators = 100, random_state = 0)
xgb.fit(X_train, y_train)
```

### Predicting on Test Data

```python
# Predicting on test data
y_pred = xgb.predict(X_test)
```

```python
y_pred
```

```python
# Creating a dataframe to compare predicted and test values
pd.DataFrame({"y_pred": y_pred, "y_test": y_test})
```

### Calculating Mean Absolute Percentage Error (MAPE)

```python
# Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
```

```python
mape
```
